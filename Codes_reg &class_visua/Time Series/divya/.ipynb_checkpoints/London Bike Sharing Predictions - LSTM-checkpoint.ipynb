**Train-test split**

```python
# train-test split shouldn't be done randomly like for other types of problems. Sequentiality should be preserved for time series data.
train_size = int(len(df) * 0.9)
test_size = len(df) - train_size
train, test = df.iloc[0:train_size], df.iloc[train_size:len(df)]
print(len(train), len(test))
```

**Data Pre-processing**

```python
# Scaling the numerical features we’re using for our modeling:
f_columns = ['t1', 't2', 'hum', 'wind_speed']
f_transformer = RobustScaler()
f_transformer = f_transformer.fit(train[f_columns].to_numpy())
train.loc[:, f_columns] = f_transformer.transform(
  train[f_columns].to_numpy()
)
test.loc[:, f_columns] = f_transformer.transform(
  test[f_columns].to_numpy()
)
```

```python
# We’ll also scale our target - the number of bike shares too:
cnt_transformer = RobustScaler()
cnt_transformer = cnt_transformer.fit(train[['cnt']])
train['cnt'] = cnt_transformer.transform(train[['cnt']])
test['cnt'] = cnt_transformer.transform(test[['cnt']])
```

Preparing sequences, using create_dataset() function:

```python
def create_dataset(X, y, time_steps=1):
    Xs, ys = [], []
    for i in range(len(X) - time_steps):
        v = X.iloc[i:(i + time_steps)].values
        Xs.append(v)
        ys.append(y.iloc[i + time_steps])
    return np.array(Xs), np.array(ys)
```

```python
time_steps = 24
# reshape to [samples, time_steps, n_features]
X_train, y_train = create_dataset(train, train.cnt, time_steps)
X_test, y_test = create_dataset(test, test.cnt, time_steps)
print(X_train.shape, y_train.shape)
```

**Model Building to Forecast demand(bike shares) using Bidirectional LSTM model**

```python
model = keras.Sequential()
model.add(
  keras.layers.Bidirectional(
    keras.layers.LSTM(
      units=128,
      input_shape=(X_train.shape[1], X_train.shape[2])
    )
  )
)
model.add(keras.layers.Dropout(rate=0.2))
model.add(keras.layers.Dense(units=1))
model.compile(loss='mean_squared_error', optimizer='adam')
```

Note: We are not going to shuffle timeseries data.

```python
history = model.fit(
    X_train, y_train,
    epochs=30,
    batch_size=32,
    validation_split=0.1,
    shuffle=False
)
```

```python
plt.plot(history.history['loss'], label='train')
plt.plot(history.history['val_loss'], label='val')
plt.legend()
```

Looks like the model learns pretty quickly. At about  5th EPOCH itself, the model is working well on validation data.

Also, observe that somewhere around 15th EPOCH is good enough after which it overfits a bit.

**Let's take predictions on test data**

```python
y_pred = model.predict(X_test)
# Reversing the transormation - Scaling
y_train_inv = cnt_transformer.inverse_transform(y_train.reshape(1,-1))
y_test_inv = cnt_transformer.inverse_transform(y_test.reshape(1,-1))
y_pred_inv = cnt_transformer.inverse_transform(y_pred)
```

 **visualizing our predictions against actual values**

```python
plt.plot(y_test_inv.flatten(), marker='.', label='Actual')
plt.plot(y_pred_inv.flatten(), 'r', marker='.', label='Predicted')
plt.legend()
```

The model is doing quite well, except that it is not able to predict some extreme values.
